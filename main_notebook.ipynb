##  (1) Free Currency Rates API

- For this I want to request closing exchange rate at the first session of the year (2023-01-02) and compare it with recent day (2023-11-20)

```python
#First requesting the close price at first trading session of 2023
fx_02012023 = requests.get("https://cdn.jsdelivr.net/gh/fawazahmed0/currency-api@1/2023-01-02/currencies/eur.json")
print("fx_02012023:", fx_02012023.status_code)
```

```python
results_02012023 = fx_02012023.json()
print(type(results_02012023))
print(len(results_02012023))
```

```python
results_02012023
```

```python
data = pd.DataFrame.from_dict(results_02012023["eur"], orient="index").reset_index()
data.columns = ["currency", "rate_2023_01_02"]
data
```

```python
#second requesting the close price at 20 November 2023
fx_20112023 = requests.get("https://cdn.jsdelivr.net/gh/fawazahmed0/currency-api@1/2023-11-20/currencies/eur.json")
print("fx_20112023:", fx_20112023.status_code)
```

```python
results_20112023 = fx_20112023.json()
```

```python
results_20112023
```

```python
data2 = pd.DataFrame.from_dict(results_20112023["eur"], orient="index").reset_index()
data2.columns = ["currency", "rate_2023_11_20"]
data2
```

```python
#merging the 2 datasets
merged_data = pd.merge(data, data2, on='currency', how='left')
```

```python
#adding year-to-date change:
merged_data['2023_YTD'] = round((merged_data['rate_2023_11_20']/merged_data['rate_2023_01_02']-1)*100,2)
```

```python
merged_data
```
